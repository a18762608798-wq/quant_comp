## 1.少量量子比特的制备方法
### 1.1电子自旋
* representation transform
电子自旋为非经典态，没有x,z方向的正交性。从z到x表象的变换为两个完备表象的幺正变换。

其变换如下

$$
|\psi\rangle_x=\frac{1}{\sqrt 2}\begin{Bmatrix}
1&1\\
1&-1
\end{Bmatrix}|\psi\rangle_z
$$

### 1.2光子的偏振

光子偏振是个相对经典的态，需要两个方向才完备。要想达到和电子自旋类似的效果，我们不可以取正交方向作为双表象。一般对于 $|0\rangle$ 取 0 rad,$|1\rangle$ 取90 rad 。 则 $|+\rangle$ 取45 rad。显然有变换关系：

$$
|+\rangle\propto |0\rangle+|1\rangle\\
|-\rangle\propto |0\rangle-|1\rangle
$$

这就和电子自旋类似了。

## 2.量子加密
* premise 

经典加密大致有两种
1. 密码本只有发送和接收两个人知道。
2. 密码本基于一些难以逆运算的加密算法，比如大数分解。

按理说第1种是绝对安全的。但除了发送者面对面写给接收者密码本，我们还要远距离传输这个密码本，并保证不被窃听。

这个就对安全需要说明：比如对于一串只有两个人知道的二进制编码，每一个子长度代表一个字母，规定A1,B1,C1,D1,...,A2,B2,....（或者按字母使用频率A1A2A3...B1B2B3...更好？但重复的问题会严重。）其中数字代表第几次出现这个字母。如果害怕重复可以取长一点或去重。
### 2.1 BB84 协议制作密码本
#### 2.1.1 Foundation
* 我们选用制备的 $|0\rangle\quad|1\rangle\quad|+\rangle\quad|-\rangle$,一般用光子偏振片，但这不重要。
* 发射者Alice生成一堆01随机数，然后随机使用一种偏振片发出去（或者说是随机选择一组基矢）。
* 接受信息者也不知道到底对面用的什么偏振片，只能胡乱用偏振片接受（理论可用一种，但这种情况就和密码写00000000一样）。
* 然后他们公开所用偏振片（当然编码不公开）。记录偏振方向一样的偏振片对应编码为有效信息，显然大约有50%有效信息。。
#### 2.1.2 被监听情况
* 假如出现Eve去拿偏振片在中间监听。它可以得到上一个情况（无监听）中Bob得到的所有信息，但是它会对信息造成坍缩破坏。
* 坍缩效果如下：
1. 如果Eve猜对了偏振方向，那没什么影响。
2. 如果Eve没有猜对，那会让量子态坍缩到这个偏振方向
* 然后Bob收到偏振片，公开相同偏振片的量子态（大概N/2）个。
* 理论上Bob这些相同偏振片的01编码是和Alice相同的。他们会公开一段编码对照一下。
* 但是如果被Eve监听过了，有约1/2的编码会被Eve猜对偏振方向不变，约1/2会坍缩到另一个量子态。这样Bob检验这些坍缩后的编码只有1/2的概率和Alice的相同。合计有约1/4的编码错误。

### 2.2 综述
综上对于密码本对照结果
1. 如果错误率在1/4左右，直接废弃掉。因为纠错也几乎没有有效信息。
2. 如果错误率很小，可以使用经典算法纠错，获得密码本。

### 2.3附录
#### 2.3.1 关于传统密码加密的一个猜想
传统密码由于位数原因不会直接作为密码本，需要扩展。

理论不太容易做可逆运算的方法都可以。我们可以拿此密码做随机种子，生成伪随机数。这是一种方法。依次伪随机数做密码本。

这种方法的缺点是穷举少量位数还是可以实现的。

而量子加密就可以直接生成密码本都没问题，甚至不放心可以一次一密。

此时哪怕穷举也会失去意义，因为会得到相同位数的任何信息。
#### 2.3.2 单光子替代问题。

单光子不大容易制备，一般使用同偏振的激光替代。

但是同偏振的激光又容易被人取下一部分又对其他光子没有影响。一般会做一些诱骗态钓鱼执法。

## 3 希尔伯特空间

即线性空间。复合系统的态适量为单个系统的态矢量的张量积。
### 3.1 所谓纠缠态
纠缠态的定义抽象：不是直积态，就是纠缠态。

那么什么是直积态呢？？

直积态的定义：

$$
\ket{\psi} = \ket{\psi_1} \otimes \ket{\psi_2}
$$

这里 $\psi_1$ 和 $\psi_2$ 一般会选取一组基矢（不是一组也可以，就是正交性头痛）。这样在希尔伯特空间量子态数量将m^n提升。m为单个量子比特状态数，n为量子比特数。我们可以把直积态展开：

$$
\ket{\psi}= a\ket{00} + b\ket{01} + c\ket{10} + d\ket{11}
$$

这个形式是通用的，但如果有的态（问题：这个态拿头制备。），不能还原为直积形式，比如：

$$
\ket{\psi} = \ket{01}-\ket{10}
$$

那么这个态就是纠缠态。我们还原不了直积形式，只需要测量第一位就一定知道第二个比特的量子态。

## 4. EPR佯谬和贝尔不等式
### 4.1 EPR佯谬
EPR 佯谬：实在性和定域性，即：

实在性：即存在上帝视角，存在基本事件（概率论基本事件都为独立事件）。可以通过基本事件的概率算出任何事件概率。注意4.2的粒子有相同表象纠缠态不同表象取反的束缚。所以只有8个基本事件。

违背实在性，意为量子力学没有基本事件，或者所谓基本事件不独立。换言之，我们的基本事件是满足约束条件所有情况的组合，但是假设Alice测量结果为 a+,Bob测量两次，分别为b,c那么由于不对易，测量顺序甚至有所影响。b,c不可表示在同一组basic vector,自然不可组为基本概率事件。

* 量子力学投影模方
对于 $P(a+,b+)$
$$
\ket{\psi}=\ket{i+}\bra{i+}{0\mu}\rangle\bra{0\mu}\psi\rangle
$$

纠缠态， $\mu=1$ 才有值。采用双比特幺正通式变化，令 $\varphi=0$
$$
primal=\ket{\psi}=\ket{i+}\bra{i+}{01}\rangle\bra{01}\psi\rangle=\\
\ket{++}\bra{++}{01}\rangle\bra{01}\psi\rangle+\ket{-+}\bra{-+}{01}\rangle\bra{01}\psi\rangle=\\
\ket{++}*[-\sin\frac{\theta}{2}\cos\frac{\theta}{2}]*\frac{1}{\sqrt 2}+\ket{-+}*[-\sin^2\frac{\theta}{2}]*\frac{1}{\sqrt 2}
$$

其模方显然为 $\frac{1}{2}\sin^2\frac{\theta}{2}$,代入 $\sigma_i$ 独立的不等式，有：
$$
\sin^2\frac{\theta_{ab}}{2}\le\sin^2\frac{\theta_{ac}}{2}+\sin^2\frac{\theta_{bc}}{2}
$$

以 $\theta_{ab}=\theta_{ac}+\theta_{bc}=2\theta_{bc}$ 为例，即证明 $\sin^2\theta\le 2\sin^2\frac{\theta}{2}$

```python
# Library
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
## Set the font globally
font_path = '/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = [FontProperties(fname=font_path).get_name()]
plt.rcParams['axes.unicode_minus'] = False  # Ensure the minus sign is displayed correctly

# Graph
## figure
theta_range=[0,np.pi/2]
x_array = np.linspace(theta_range[0],theta_range[1],1000)
y0_array = np.sin(x_array)**2
y1_array = 2*np.sin(x_array/2)**2
plt.plot(x_array, y0_array, label=r'$\sin^2\theta$')
plt.plot(x_array, y1_array, label=r'$2\sin^2{\frac{\theta}{2}}$')
plt.legend(loc='upper left')
plt.title('图表标题')
plt.xlabel('横坐标')
plt.ylabel('纵坐标')
plt.show()
```
