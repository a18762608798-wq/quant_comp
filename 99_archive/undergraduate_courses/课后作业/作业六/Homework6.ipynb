## $\S$ 0、预处理
* 定义二进制函数

对于二进制整数： $j=j_1j_2...j_i.j_k...j_n$

求对应十进制值以做运算。

```python
import sympy as sp
import numpy as np
from numba import jit 
from IPython.display import display
# funtion
def deci_bin(integ:np.array,frac:np.array):
    num_integ=np.size(integ)
    num_frac=np.size(frac)
    deci=0
    for i in range(num_integ):
        deci+=integ[-i-1]*2**i
    for i in range(num_frac):
        deci+=frac[i]/2**(i+1)
    return deci
# test
integ=[1,1]
frac=[0,0]
deci_bin(integ,frac)
```

* 双比特傅里叶变换系数

```python
def coef(j1,j2,k1,k2):
    return 1/2*sp.exp(2*sp.pi*sp.I*(2*j1+j2)*(2*k1+k2)/4)
```

## $\S 1$ 问题一

```python
# 符号变量设置
j1,j2= sp.symbols('j_1 j_2',integer=True)
integ=[j1]
frac=[j2]
# 计算
## 中间量
expr1=deci_bin(integ,frac)
expr2=deci_bin([0],frac)
## 结果
left_=sp.exp(2*sp.pi*sp.I*expr1)
right_=sp.exp(2*sp.pi*sp.I*expr2)
# 输出
print("等式两边分别为：")
display(left_,right_)
print("是否相等判断：")
display(sp.simplify(left_)==sp.simplify(right_))
```

## $\S 2$、问题二

### 2.0 证明

```python
# 计算系数
coef_cal=sp.simplify(sp.Matrix([\
coef(j1,j2,0,0),
coef(j1,j2,0,1),
coef(j1,j2,1,0),
coef(j1,j2,1,1),
]))

# 题目中的系数
## 中间量
### 第2项
integ=[0]
frac=[j1,j2]
expr2=deci_bin(integ,frac)
### 第3项
integ=[j1]
frac=[j2]
expr3=deci_bin(integ,frac)
### 系数向量
coef_kown=sp.simplify(sp.Matrix([\
1/2,
1/2*sp.exp(2*sp.pi*sp.I*expr2),
1/2*sp.exp(2*sp.pi*sp.I*expr3),
1/2*sp.exp(2*sp.pi*sp.I*(expr2+expr3)),
]))

# display
print("等式两边分别为：")
display(coef_cal,coef_kown)
print("是否相等判断：")
display(coef_cal==coef_kown)
```
