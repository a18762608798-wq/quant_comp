## question 1

```python
from sympy.physics.quantum.qubit import Qubit
from sympy.physics.quantum import gate
from sympy.physics.quantum import represent
from sympy.physics.quantum.tensorproduct import TensorProduct
import sympy as sp

# 定义
## 定义量子态
ket00 =Qubit('00')
ket01 =Qubit('01')
ket10 =Qubit('10')
ket11 =Qubit('11')
a00,a01,a10,a11,phi = sp.symbols('a00 a01 a10 a11 phi')
phi=a00*ket00+a01*ket01+a10*ket10+a11*ket11
## 定义量子门
CNOT_gate = gate.CNOT(0, 1)*gate.CNOT(1, 0)*gate.CNOT(0, 1)

# 计算
## 量子比特作用
result = CNOT_gate * phi

# display
print("初态：")
display(represent(phi , nqubits=2)) 
print("总作用算符：")
display(represent(CNOT_gate , nqubits=2)) 
print("末态：")
display(represent(result , nqubits=2)) 
```

## question2

这三个CNOT_gate作用显然是交换 $|10\rangle$ 和 $|01\rangle$ 的系数。

但是看不出效果，应该看单个比特上的差别。

$$
\begin{cases}
\psi_0=a_0|0\rangle+a_1|1\rangle\\
\psi_1=b_0|0\rangle+b_1|1\rangle\\
\end{cases}
$$

```python
# 定义
a0,a1,b0,b1= sp.symbols('a0 a1 b0 b1')
phi0=a0*Qubit('0')+a1*Qubit('1')
phi1=b0*Qubit('0')+b1*Qubit('1')
phia=phi1
phib=phi0
phi=TensorProduct(phi0,phi1)
phi_=TensorProduct(phia,phib)

# 计算
## 量子比特作用
result = sp.simplify(CNOT_gate * phi)
result_ = sp.simplify(CNOT_gate * phi_)
# display
print("初态1")
display(represent(phi0 , nqubits=1))
display(represent(phi1 , nqubits=1))
print("对比态")
display(represent(phia , nqubits=1))
display(represent(phib , nqubits=1))
print("初态对比态")
display(represent(phi_ , nqubits=2))
print("末态1")
represent(result , nqubits=2)
```

末态1等于对比态，也即是把两个比特量子态交换
